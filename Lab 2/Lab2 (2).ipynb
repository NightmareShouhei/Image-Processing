```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
def sample_image(image, factor):
    """
    Downsamples the image by the given factor.
    Args:
        image (numpy array): Original image.
        factor (int): Factor by which to downsample.
    Returns:
        numpy array: Downsampled image.
    """
    height, width = image.shape[:2]
    sampled_image = cv2.resize(image, (width // factor, height // factor), interpolation=cv2.INTER_NEAREST)
    return sampled_image

def quantize_image(image, levels):
    """
    Reduces the number of grayscale levels in the image.
    Args:
        image (numpy array): Original image.
        levels (int): Number of grayscale levels.
    Returns:
        numpy array: Quantized image.
    """
    quantized_image = np.floor(image / (256 // levels)) * (256 // levels)
    quantized_image = quantized_image.astype(np.uint8)
    return quantized_image

def plot_images(original, sampled, quantized):
    """
    Plots the original, sampled, and quantized images side by side.
    Args:
        original (numpy array): Original image.
        sampled (numpy array): Sampled image.
        quantized (numpy array): Quantized image.
    """
    plt.figure(figsize=(12, 4))
    plt.subplot(1, 3, 1)
    plt.imshow(original, cmap='gray')
    plt.title('Original Image')
    plt.axis('off')

    plt.subplot(1, 3, 2)
    plt.imshow(sampled, cmap='gray')
    plt.title('Sampled Image')
    plt.axis('off')

    plt.subplot(1, 3, 3)
    plt.imshow(quantized, cmap='gray')
    plt.title('Quantized Image')
    plt.axis('off')

    plt.show()
```

```python
def original_sample_quantized(image, sampling_factor, quantization_levels):

    # Sample and quantize
    sampled_image = sample_image(image, sampling_factor)
    quantized_image = quantize_image(image, quantization_levels)

    # Plot results
    plot_images(image, sampled_image, quantized_image)
```

```python
# Load images
image_a = cv2.imread('Images/image_a.png')
image_b = cv2.imread('Images/image_b.png')
image_c = cv2.imread('Images/cameraman.tif')

# Resize
resize = (400, 400)
image_a = cv2.resize(image_a, resize, interpolation=cv2.INTER_LANCZOS4)
image_b = cv2.resize(image_b, resize, interpolation=cv2.INTER_LANCZOS4)
image_c = cv2.resize(image_c, resize, interpolation=cv2.INTER_LANCZOS4)

# To array
imgArr1 = np.asarray(image_a)
imgArr2 = np.asarray(image_b)
```

```python
# sampling and quantization
original_sample_quantized(image_c, 14, 9)
original_sample_quantized(image_c, 5, 2)
original_sample_quantized(image_c, 1, 15)
original_sample_quantized(image_c, 30, 1)
```

```python
# Subtract two images and display the result
result = cv2.subtract(imgArr1, imgArr2)
resultImage = Image.fromarray(result)
resultImage.show()
```

```python
# Add one image with a constant value of 175 and display it.
result = cv2.add(imgArr1, 175)
resultImage = Image.fromarray(result)
resultImage.show()
```

```python
# Apply the set difference operation
result = imgArr1 & (~imgArr2)
resultImage = Image.fromarray(result)
resultImage.show()
```

```python
# Apply the symmetric difference operation
result = imgArr1 ^ imgArr2
resultImage = Image.fromarray(result)
resultImage.show()
```

```python
# Apply Intersection operation
result = imgArr1 & imgArr2
resultImage = Image.fromarray(result)
resultImage.show()
```
