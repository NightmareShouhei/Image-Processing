```python
#Hamza ALsihati
#2230006810
#8MA1
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage import data
```

```python
# Task 1
image = data.chelsea()
smoothed = cv2.boxFilter(image, ddepth=-1, ksize=(7, 7))
fig, ax = plt.subplots(1, 2, figsize=(10, 5))

ax[0].imshow(image)
ax[0].set_title("Original")
ax[0].axis("off")

ax[1].imshow(smoothed)
ax[1].set_title("Smoothed")
ax[1].axis("off")

plt.tight_layout()
plt.show()
```

```python
# Task 2
image2 = data.chelsea()

gauss_small = cv2.GaussianBlur(image2, (5, 5), sigmaX=0)
gauss_large = cv2.GaussianBlur(image2, (21, 21), sigmaX=0)

fig, ax = plt.subplots(1, 3, figsize=(15, 5))
ax[0].imshow(image2)
ax[0].set_title("Original")
ax[0].axis("off")

ax[1].imshow(gauss_small)
ax[1].set_title("5x5 Gaussian")
ax[1].axis("off")

ax[2].imshow(gauss_large)
ax[2].set_title("21x21 Gaussian")
ax[2].axis("off")

plt.tight_layout()
plt.show()
```

```python
# Task 3
image3 = data.chelsea()

image_float = image3.astype(np.float32) / 255.0
laplacian = cv2.Laplacian(image_float, cv2.CV_32F, ksize=3)
sharpened = np.clip(image_float - laplacian, 0, 1)

fig, ax = plt.subplots(1, 3, figsize=(15, 5))
ax[0].imshow(image_float)
ax[0].set_title("Original")
ax[0].axis("off")

ax[1].imshow(np.clip(laplacian, 0, 1))
ax[1].set_title("Laplacian Filter")
ax[1].axis("off")

ax[2].imshow(sharpened)
ax[2].set_title("Sharpened")
ax[2].axis("off")

plt.tight_layout()
plt.show()
```
